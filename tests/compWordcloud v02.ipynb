```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import random
```

```python
import json

data = json.load(open('data/kwrd_bcp.json'))
```

```python
pallete = ["#1671d2", "#e39f46", "#636363", "#b0bec5"]
seed = 0
        


word_freq = {word: data1["weight"] for word, data1 in data.items() if data1["filter"] >= 0.1}

def color_function(word, font_size, position, orientation, random_state=None, **kwargs):
    global pallete
    return random.choice(pallete)

wordcloud = WordCloud(width=800, height=400, background_color=None, mode='RGBA').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
random.seed(seed)
plt.imshow(wordcloud.recolor(color_func=color_function), interpolation='nearest')
plt.axis('off')

plt.show()
```
