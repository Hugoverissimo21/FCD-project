# SAVE THE KEYWORDS TO A JSON

```python
import json
import random

with open('data/kwrd_bcp.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)

data
```

```python
keywords = list(data.keys())
```

```python
def hilo(data):
    keywords = list(data.keys())

    word1 = random.choice(keywords)
    word1_mentions = int(data[word1]["count"])

    word2 = random.choice(keywords)
    word2_mentions = int(data[word2]["count"])

    score = 0
    while True:
        print(word1, word1_mentions)
        print(word2)

        ans = input(f"Does {word2} have more mentions than {word1}? (y/n) ")

        scenarios = {
            "y": word2_mentions > word1_mentions,
            "n": word2_mentions < word1_mentions
        }

        if scenarios[ans]:
            score += 1
            print(f"Correct! Score: {score}.")
            word1 = word2
            word1_mentions = word2_mentions
            word2 = random.choice(keywords)
            word2_mentions = int(data[word2]["count"])
        
        else:
            print(f"Incorrect! Final score: {score}.")
            break

hilo(data)
```
