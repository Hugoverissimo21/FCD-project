```python
from matplotlib_venn import venn3_circles
from matplotlib_venn import venn3

lista1 = {"maçã", "banana", "uva", "laranja", "pêra"}
lista2 = {"uva", "laranja", "melancia", "abacaxi", "banana"}
lista3 = {"pêra", "melancia", "kiwi", "uva", "caju"}
lista4 = {"banana", "pêssego", "laranja", "manga", "caju"}
lista5 = {"maçã", "melancia", "caju", "abacaxi", "pêra"}

intersecao_12 = lista1 & lista2
intersecao_13 = lista1 & lista3
intersecao_23 = lista2 & lista3
comum_123 = lista1 & lista2 & lista3


# Preparando as palavras para exibir nas áreas correspondentes
venn_labels = {
    '100': "\n".join(lista1 - (lista2 | lista3)),
    '010': "\n".join(lista2 - (lista1 | lista3)),
    '001': "\n".join(lista3 - (lista1 | lista2)),
    '110': "\n".join(intersecao_12 - comum_123),
    '101': "\n".join(intersecao_13 - comum_123),
    '011': "\n".join(intersecao_23 - comum_123),
    '111': "\n".join(comum_123),
}

# Criando o diagrama com palavras
plt.figure(figsize=(10, 6))
venn = venn3(subsets=(len(lista1), len(lista2), len(intersecao_12),
                      len(lista3), len(intersecao_13), len(intersecao_23),
                      len(comum_123)), set_labels=('Lista 1', 'Lista 2', 'Lista 3'))

# Atualizando cada área com as palavras
for idx, label in venn_labels.items():
    venn.get_label_by_id(idx).set_text(label)

# Exibindo o gráfico com as palavras
plt.title("Diagrama de Venn com Palavras nas Interseções")
plt.show()
```

```python
from itertools import combinations

# Função para calcular interseções e convertê-las em string
def calcular_intersecoes(sets):
    intersecoes = {}
    for i in range(1, len(sets) + 1):
        for combo in combinations(range(len(sets)), i):
            intersecao = set.intersection(*[sets[j] for j in combo])
            key = ''.join(['1' if k in combo else '0' for k in range(len(sets))])
            intersecoes[key] = "\n".join(map(str, intersecao))
    return intersecoes

# Exemplo de uso com listas
lista1 = {"maçã", "banana", "uva", "laranja", "pêra"}
lista2 = {"uva", "laranja", "melancia", "abacaxi", "banana"}
lista3 = {"pêra", "melancia", "kiwi", "uva", "caju"}

# Calculando interseções
listas = [lista1, lista2, lista3]
venn_labels = calcular_intersecoes(listas)

# Exibindo as interseções calculadas automaticamente
print(venn_labels)
```

```python
from upsetplot import UpSet
import pandas as pd

# Definindo cinco listas de palavras
lista1 = {"maçã", "banana", "uva", "laranja", "pêra"}
lista2 = {"uva", "laranja", "melancia", "abacaxi", "banana"}
lista3 = {"pêra", "melancia", "kiwi", "uva", "caju"}
lista4 = {"banana", "pêssego", "laranja", "manga", "caju"}
lista5 = {"maçã", "melancia", "caju", "abacaxi", "pêra"}

# Colocando as listas em um dicionário
listas = {"Lista1": lista1, "Lista2": lista2, "Lista3": lista3, "Lista4": lista4, "Lista5": lista5}

# Preparando os dados para o gráfico UpSet
# Usamos um MultiIndex que diferencia cada item presente nas listas
intersections = {}
for name, group in listas.items():
    for item in group:
        key = tuple(item in listas[list_name] for list_name in listas)
        intersections[key] = intersections.get(key, 0) + 1

# Convertendo para uma série pandas com MultiIndex
index = pd.MultiIndex.from_tuples(intersections.keys(), names=listas.keys())
intersections = pd.Series(list(intersections.values()), index=index)

# Criando o gráfico UpSet
upset = UpSet(intersections)
upset.plot()
plt.suptitle("Interseções entre cinco listas de palavras")
plt.show()
```
