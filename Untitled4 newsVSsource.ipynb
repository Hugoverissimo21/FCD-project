```python
def getSource(x):
    sources = {}
    for i in x:
        try:
            source = i["newsSource"]
            if source in sources:
                sources[source] += 1
            else:
                sources[source] = 1
        except:
            return {}
    return sources


df = df.map(getSource)
```

```python
import matplotlib.pyplot as plt
import numpy as np


df.index = df.index

# Extract unique news sources
all_sources = set()
for i in df.index:
    all_sources.update(df["news"].loc[i].keys())

all_sources = list(all_sources)  # Convert to list

# Set number of categories and x locations
n_categories = len(df.index)
n_sources = len(all_sources)
x = np.arange(n_sources)

# Set width of bars
width = 0.1  # Adjust the width as needed

# Create the bar plot
plt.figure(figsize=(12, 6))

# Loop over each index and plot bars
for i, idx in enumerate(df.index):
    counts = [df["news"].loc[idx].get(source, 0) for source in all_sources]  # Get counts for the current index
    plt.bar(x + i * width, counts, width, label=idx)  # Offset x position by i * width

# Add labels and title
plt.xlabel('News Sources')
plt.ylabel('Count')
plt.title('news by news source')
plt.xticks(x + width*2, all_sources)  # Set x-ticks to news sources
plt.legend()  # Show legend

# Adjust layout for better fit
plt.tight_layout()

# Show the plot
plt.show()
```
