```python
from itertools import combinations

# Função para calcular interseções e convertê-las em string
def calcular_intersecoes(sets):
    intersecoes = {}
    for i in range(1, len(sets) + 1):
        for combo in combinations(range(len(sets)), i):
            intersecao = set.intersection(*[sets[j] for j in combo])
            key = ''.join(['1' if k in combo else '0' for k in range(len(sets))])
            intersecoes[key] = "\n".join(map(str, intersecao))
    return intersecoes

# Exemplo de uso com listas
lista1 = {"maçã", "banana", "uva", "laranja", "pêra"}
lista2 = {"uva", "laranja", "melancia", "abacaxi", "banana"}
lista3 = {"pêra", "melancia", "kiwi", "uva", "caju"}

# Calculando interseções
listas = [lista1, lista2, lista3]
venn_labels = calcular_intersecoes(listas)

# Exibindo as interseções calculadas automaticamente
print(venn_labels)
```
